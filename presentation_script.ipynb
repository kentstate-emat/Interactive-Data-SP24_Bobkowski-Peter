```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
cricket = pd.read_csv('cricket.csv')
```

```python
cricket
```

```python
fig = px.scatter(cricket,\
                 x = 'Innings',\
                 y = 'Runs',\
                 height = 500,\
                 title = "Cricket Player Statistics")
fig.show()
```

```python
fig = px.scatter(cricket,\
                 x="Innings",\
                 y="Runs",\
                 height = 500,\
                 title = "Cricket Player Statistics",\
                 hover_name="Player",\
                 hover_data=["Average", "Country"])
fig.show()
```

```python
# Button that changes scatterplot to bar graph

fig = px.scatter(cricket,\
                 x = "Innings",\
                 y = "Runs",\
                 hover_name = "Player",\
                 hover_data = ["Average", "Country"],\
                 height = 500,\
                 title = "Cricket Player Statistics" )

df_india = cricket[cricket['Country'] == 'INDIA']

fig.update_layout(
    updatemenus=[
        dict(
            type = "buttons",
            buttons = list([
                dict(
                    args=["type", "scatter"],
                    label="Scatter Plot",
                    method="restyle"
                ),
                dict(
                    args=["type", "bar"],
                    label="Bar Chart",
                    method="restyle"
                )
            ]),
            direction = "left",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1,
            yanchor="top"
        ),
    ]
)


fig.show()
```

```python
# Button changes y-axis from "Runs" to "No Outs"
# But the y-axis label doesn't change :(

fig = px.scatter(cricket,\
                 x = "Innings",\
                 y = "Runs",\
                 hover_name = "Player",\
                 hover_data = ["Average", "Country"],\
                 height = 500,\
                 title = "Cricket Player Statistics")

fig.update_layout(
    updatemenus=[
        dict(
            type = "buttons",
            buttons = list([
                dict(
                    args = [{'y': [cricket['Runs']],
                             'yaxis_title': 'Runs'
                            }],
                    label = "Runs",
                    method = "update"
                ),
                dict(
                    args = [{'y': [cricket['No Outs']],
                             'yaxis_title': 'No Outs'
                            }],
                    label = "No Outs",
                    method = "update"
                )
            ]),
            direction = "left",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1,
            yanchor="top"
        ),
    ]
)


fig.show()
```

```python
# Button that changes display from all countries to India

fig = px.scatter(cricket,\
                 x = "Innings",\
                 y = "Runs",\
                 hover_name = "Player",\
                 hover_data = ["Average", "Country"],\
                 height = 500,\
                 title = "Cricket Player Statistics" )

df_india = cricket[cricket['Country'] == 'INDIA']

fig.update_layout(
    updatemenus=[
        dict(
            type = "buttons",
            direction = "left",
            buttons = list([
                dict(
                    args = [{'x': [cricket['Innings']], 
                             'y': [cricket['Runs']]
                            }],
                    label = "All",
                    method = "update"
                ),
                dict(
                    args = [{'x': [df_india['Innings']], 
                             'y': [df_india['Runs']]
                            }],
                    label = "India",
                    method = "update"
                )
            ]),
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1,
            yanchor="top"
        ),
    ]
)


fig.show()
```

```python
fig2 = px.scatter(cricket,\
                  x = "Innings",\
                  y = "Runs",\
                  hover_name = "Player",\
                  hover_data = ["Average", "Country"])

fig2.update_layout(
    width=800,
    height=1000,
    autosize=False,
    margin=dict(t=200, b=0, l=0, r=0),
)

df_india = cricket[cricket['Country'] == 'INDIA']

# button_layer_1_height = 1.08
button_layer_1_height = 1.12
button_layer_2_height = 1.065

fig2.update_layout(
    updatemenus=[
        dict(
            buttons=list([
                dict(
                    args = [{'x': [cricket['Innings']], 
                             'y': [cricket['Runs']]
                            }],
                    label = "All",
                    method = "update"
                ),
                dict(
                    args = [{'x': [df_india['Innings']], 
                             'y': [df_india['Runs']]
                            }],
                    label = "India",
                    method = "update"
                )
            ]),
            type = "buttons",
            direction="right",
            pad={"r": 9, "t": 9},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=button_layer_1_height,
            yanchor="top"
        ),
        dict(
            buttons=list([
               dict(
                    args=["type", "scatter"],
                    label="Scatter Plot",
                    method="restyle"
                ),
                dict(
                    args=["type", "bar"],
                    label="Bar Chart",
                    method="restyle"
                )
            ]),
            type = "buttons",
            direction="right",
            pad={"r": 9, "t": 9},
            showactive=True,
            x=0.13,
            xanchor="left",
            y=button_layer_2_height,
            yanchor="top"
        ),
    ])

fig2.update_layout(
    annotations=[
        dict(text="Countries", x=0, xref="paper", y=1.1, yref="paper",
                             align="left", showarrow=False),
        dict(text="Chart<br>Type", x=0, xref="paper", y=1.06,
                             yref="paper", showarrow=False),
    ])

fig2.show()
```

```python
fig1 = px.scatter(cricket,\
                  x = "Innings",\
                  y = "Runs",\
                  hover_name = "Player",\
                  # labels = {'Runs': 'Runs'},
                  hover_data = ["Average", "Country"],\
                  height = 500,\
                  title = "Cricket Player Statistics")

button1 = {'method': 'update',
          'label': 'Runs',
          'args': [
              {'y': [cricket['Runs']]},
              {'yaxis.title': 'Runs'}]
          }

button2 = {'method': 'update',
          'label': 'No Outs',
          'args': [
              {'y': [cricket['No Outs']]},
              {'yaxis.title': 'No Outs'}]
          }

fig1.update_layout(
    updatemenus=[{
        'type': 'buttons',
        'direction': "left",
        'pad': {"r": 10, "t": 10},
        'showactive': True,
        'x': 0.1,
        'xanchor': "left",
        'y': 1,
        'yanchor': "top",
        'buttons': [button1, button2]
    }]
)

fig1.show()
```
