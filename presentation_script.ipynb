```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
cricket = pd.read_csv('cricket.csv')
```

```python
cricket
```

```python
fig = px.scatter(cricket,\
                 x = 'Innings',\
                 y = 'Runs',\
                 height = 500,\
                 title = "Cricket Player Statistics")
fig.show()
```

```python
fig = px.scatter(cricket,\
                 x="Innings",\
                 y="Runs",\
                 height = 500,\
                 title = "Cricket Player Statistics",\
                 hover_name="Player",\
                 hover_data=["Average", "Country"])
fig.show()
```

```python
# Button that changes scatterplot to bar graph

fig = px.scatter(cricket,\
                 x = "Innings",\
                 y = "Runs",\
                 hover_name = "Player",\
                 hover_data = ["Average", "Country"],\
                 height = 500,\
                 title = "Cricket Player Statistics" )

df_india = cricket[cricket['Country'] == 'INDIA']

fig.update_layout(
    updatemenus=[
        dict(
            type = "buttons",
            buttons = list([
                dict(
                    args=["type", "scatter"],
                    label="Scatter Plot",
                    method="restyle"
                ),
                dict(
                    args=["type", "bar"],
                    label="Bar Chart",
                    method="restyle"
                )
            ]),
            direction = "left",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1,
            yanchor="top"
        ),
    ]
)


fig.show()
```

```python
# Button changes y-axis from "Runs" to "No Outs"
# But the y-axis label doesn't change :(

fig = px.scatter(cricket,\
                 x = "Innings",\
                 y = "Runs",\
                 hover_name = "Player",\
                 hover_data = ["Average", "Country"],\
                 height = 500,\
                 title = "Cricket Player Statistics")

fig.update_layout(
    updatemenus=[
        dict(
            type = "buttons",
            buttons = list([
                dict(
                    args = [{'x': [cricket['Innings']], 
                             'y': [cricket['Runs']],
                             'yaxis_title': 'Runs'
                            }],
                    label = "Runs",
                    method = "update"
                ),
                dict(
                    args = [{'x': [cricket['Innings']], 
                             'y': [cricket['No Outs']],
                             'yaxis_title': 'No Outs'
                            }],
                    label = "No Outs",
                    method = "update"
                )
            ]),
            direction = "left",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1,
            yanchor="top"
        ),
    ]
)


fig.show()
```

```python
# Button that changes display from all countries to India

fig = px.scatter(cricket,\
                 x = "Innings",\
                 y = "Runs",\
                 hover_name = "Player",\
                 hover_data = ["Average", "Country"],\
                 height = 500,\
                 title = "Cricket Player Statistics" )

df_india = cricket[cricket['Country'] == 'INDIA']

fig.update_layout(
    updatemenus=[
        dict(
            type = "buttons",
            direction = "left",
            buttons = list([
                dict(
                    args = [{'x': [cricket['Innings']], 
                             'y': [cricket['Runs']]
                            }],
                    label = "All",
                    method = "update"
                ),
                dict(
                    args = [{'x': [df_india['Innings']], 
                             'y': [df_india['Runs']]
                            }],
                    label = "India",
                    method = "update"
                )
            ]),
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1,
            yanchor="top"
        ),
    ]
)


fig.show()
```
