#### **Explain how different relplot methods can be used to understand relationships**

Relplot allows the assignment of a series of facets and styles to several variables at once, letting analysts examine a dataset's complex characteristics in one plot. Relplot makes it easy to visualize both continuous and categorical data, and combinations of the two. 

The scatterplot default is useful when examining relationships between two variables where seeing individual observations is informative. The line plot works well when one of the variables represents time or another continuous variable like that. The ability to display two or more axes next to each other by the levels of a categorical variable makes it easy to compare the relationship between two or more variables as a function of another variable.

### **Group 2 Individual Challenge**

#### **Exploring other Seaborn plot types**

***Violin plot***

```python
sns.violinplot(x = "Education Level", \
               y = "Salary", \
               hue = "Gender", \
               split = True, \
               gap = 0.1, \
               palette = "Set3", \
               linewidth = 0.8, \
               data = Salary_data)
sns.despine() 
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), title='Gender', frameon=False)

plt.show()
```

***Regression plot***

```python
sns.regplot(x = "Age", \
            y = "Salary", \
            ci = 95, \
            data = Salary_data)

plt.show()
```

***Separate regression plots for females and males***

```python
from matplotlib import ticker as tkr
```

```python
# Setting up the two axes
fig1, ax1 = plt.subplots(1, 2, figsize = (10,5))
fig1.tight_layout(pad = 4.0)
ax1[0].set_title("Females")
ax1[1].set_title("Males")

# Filtering data by gender
male = Salary_data.loc[Salary_data['Gender'] == 'Male']
female = Salary_data.loc[Salary_data['Gender'] == 'Female']

fem_plot = sns.regplot(ax = ax1[0], \
                       x = "Age", \
                       y = "Salary", \
                       ci = 95, \
                       data = female)
male_plot = sns.regplot(ax = ax1[1], \
                        x = "Age", \
                        y = "Salary", \
                        ci = 95, \
                        data = male)

# Setting the axes to be equal
fem_plot.set_xlim(20, 55)
male_plot.set_xlim(20, 55)

fem_plot.set_ylim(0, 275000)
male_plot.set_ylim(0, 275000)

# Formatting y-axis tik marks (Source: https://napsterinblue.github.io/notes/python/viz/tick_string_formatting/)
def thousands(x, pos):
    return '${:,.0f}K'.format(x*1e-3)
fmtr = tkr.FuncFormatter(thousands)

fem_plot.yaxis.set_major_formatter(fmtr)
male_plot.yaxis.set_major_formatter(fmtr)

plt.show()
```
